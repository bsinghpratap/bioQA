```python

#
import torch
import json
import numpy as np
import transformers
import pandas as pd
import pickle as pkl
from torch import nn
from tqdm import tqdm
from os.path import join
from importlib import reload
import multiprocessing as mp
from collections import Counter
from data_pub import pubmedDataset
from torch.nn import CrossEntropyLoss
from torch.utils.data import DataLoader
from copy import deepcopy
from sklearn.metrics import classification_report, confusion_matrix
from transformers import (BertPreTrainedModel, BertModel, AdamW, get_linear_schedule_with_warmup, 
                          RobertaPreTrainedModel, RobertaModel,
                          AutoTokenizer, AutoModel, AutoConfig)
from transformers import (WEIGHTS_NAME,
                          AutoModelForSequenceClassification,
                          BertConfig, BertForSequenceClassification, BertTokenizer,
                          XLMConfig, XLMForSequenceClassification, XLMTokenizer,
                          DistilBertConfig, DistilBertForSequenceClassification, DistilBertTokenizer,
                          RobertaConfig, RobertaForSequenceClassification, RobertaTokenizer)

from sklearn.decomposition import PCA

#
from PubMedQAData import QADataLoader
```

```python
# hyperpars
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
args = {
    'weight_decay': 10,
    'learning_rate': 6e-6,
    'epochs': 100,
    'eval_every_steps': 10,
    'gradient_accumulation_steps': 1,
    'adam_epsilon': 1e-8,
    'max_sequence_length': 512,
    'batch_size': 64,
    'scheduler_warmup': 0.05,
    'training_phase': 'phase1',
}
label2id = {'yes': 0, 'no': 1, 'maybe': 2}

# model specs
model_dict = {
    0: {
        'model': 'roberta-large',
        'tokenizer': 'roberta-large',
    },
}

# load data
dataloaders = QADataLoader(
    datasets_name='pubmed_qa',
    datasets_config='pqa_labeled',
    label2id=label2id,
    tokenizer_name=model_dict[0]['tokenizer'],
    max_sequence_length=args['max_sequence_length'],
    batch_size=args['batch_size'],
    debug=False
)

#
train_loader = dataloaders.dataloader_train
val_loader = dataloaders.dataloader_validation
test_loader = dataloaders.dataloader_test
```

```python
# model class to get encodings
class QAModel(nn.Module):
    def __init__(
        self,
        model_name,
        num_classes,
    ):
        super(QAModel, self).__init__()

        config = AutoConfig.from_pretrained(
            model_name,
            num_labels=num_classes,
            finetuning_task='pubmedqa'
        )
        self.encoder = AutoModel.from_pretrained(
            model_name, 
            config=config,
        )
    
        return

    def forward(
        self,
        batch_,
    ):
        batch_['return_dict'] = True
        outputs = self.encoder(**batch_)
        #print(outputs['last_hidden_state'].shape)
        vec = outputs['last_hidden_state'][:, 0, :]
        #print(vec.shape)
        
        return vec
```

```python
#
def reduce_dim(X_in, to_=2):
    
    pca = PCA(n_components=2)
    pca = pca.fit(X_in)
    X_out = np.array(pca.transform(X_in))
    
    return X_out

def get_vec(model, data_loader):
    model.eval()
    vecs = {}
    eval_loss = 0
    with torch.no_grad():
        for batch_idx, batch in tqdm(enumerate(data_loader)):
            
            # unroll features
            input_batch = {
                'input_ids':batch['input_ids'],
                'attention_mask':batch['attention_mask']
            }
            input_batch = {k: v.to(device) for k, v in input_batch.items()}
            
            # forward pass
            batch_vec = model(input_batch)
            
            # collect data
            batch_vec = batch_vec.detach().cpu().numpy()
            batch_vec_pca = reduce_dim(batch_vec)
            for idx_, id_ in enumerate(batch['ids']):
                vecs[id_] = {
                    'vector': batch_vec[idx_, :],
                    'vector_pca': batch_vec_pca[idx_, :],
                    'label': batch['encoder_labels'][idx_]
                }
    
    return vecs
```

```python
#
m_ = QAModel(
    model_name=model_dict[0]['model'],
    num_classes=dataloaders.num_classes,
)
m_.to(device)

#
train_vec = get_vec(m_, train_loader)
val_vec = get_vec(m_, val_loader)
test_vec = get_vec(m_, test_loader)    
```

```python
# restructure for plotting
reps = pd.DataFrame(
    -100,
    index=range(len(train_vec) + len(val_vec) + len(test_vec)),
    columns=['example_id', 'Dim1', 'Dim2', 'label', 'subset'],
)

#
row = -1
for id_ in train_vec:
    row += 1
    
    #
    reps.loc[row, 'example_id'] = id_.numpy().tolist()
    reps.loc[row, ['Dim1', 'Dim2']] = train_vec[id_]['vector_pca'].tolist()
    reps.loc[row, 'label'] = train_vec[id_]['label'].tolist()
    reps.loc[row, 'subset'] = 'train'

for id_ in val_vec:
    row += 1
    
    #
    reps.loc[row, 'example_id'] = id_.numpy().tolist()
    reps.loc[row, ['Dim1', 'Dim2']] = val_vec[id_]['vector_pca'].tolist()
    reps.loc[row, 'label'] = val_vec[id_]['label'].tolist()
    reps.loc[row, 'subset'] = 'val.'

for id_ in test_vec:
    row += 1
    
    #
    reps.loc[row, 'example_id'] = id_.numpy().tolist()
    reps.loc[row, ['Dim1', 'Dim2']] = test_vec[id_]['vector_pca'].tolist()
    reps.loc[row, 'label'] = test_vec[id_]['label'].tolist()
    reps.loc[row, 'subset'] = 'test'

#
reps['COMP1'] = reps['Dim1'].values
reps['COMP2'] = reps['Dim2'].values
```

```python
#
data_text = {}
with open('train_set.json', 'r') as f:
    data_text['train'] = json.load(f)
with open('dev_set.json', 'r') as f:
    data_text['validation'] = json.load(f)
with open('test_set.json', 'r') as f:
    data_text['test'] = json.load(f)

data = {}
for subset in data_text:
    data[subset] = {}
    for label in ['yes', 'no', 'maybe']:
        data[subset][label] = []
    
    #
    for id_ in tqdm(data_text[subset]):
        instance = data_text[subset][id_]
        l_ = instance['final_decision']
        instance_ = {
            'question': instance['QUESTION'],
            'context': ' '.join(instance['CONTEXTS']),
            'long_answer': instance['LONG_ANSWER'],
            'short_answer': l_,
            'id': id_,
        }
        
        #
        data[subset][l_].append(instance_)
        
```
